```python

def make_feature(df):
    #血脂特征
    df["高低密度蛋白胆固醇比"]=df["高密度脂蛋白胆固醇"]/df["低密度脂蛋白胆固醇"]
    df["高密度/总胆固醇比"]=df["高密度脂蛋白胆固醇"]/df["总胆固醇"]
    df["低密度/总胆固醇比"]=df["低密度脂蛋白胆固醇"]/df["总胆固醇"]
    df["白蛋白/总蛋白比"]=df["白蛋白"]/df["*总蛋白"]
    df["球蛋白/总蛋白比"]=df["*球蛋白"]/df["*总蛋白"]
    df["白蛋白/球蛋白"]=df["白蛋白"]/df["*球蛋白"]
    #尿素特征
    df["尿素肌酐比"]=df["尿素"]/df["肌酐"]
    #红细胞，白细胞，血小板三大细胞比例
    df["红细胞比容"]=df["红细胞计数"]/(df["红细胞计数"]+df["血小板计数"]+df["白细胞计数"])
    df["白细胞比容"]=df["白细胞计数"]/(df["红细胞计数"]+df["血小板计数"]+df["白细胞计数"])
    df["血小板比容"]=df["血小板计数"]/(df["红细胞计数"]+df["血小板计数"]+df["白细胞计数"])
    df["红白细胞比"]=df["红细胞计数"]/df["白细胞计数"]
    df["红血细胞比"]=df["红细胞计数"]/df["血小板计数"]
    df["白血细胞比"]=df["白细胞计数"]/df["血小板计数"]
    #白细胞内各大细胞比例
    df["白中细胞比"]=df["中性粒细胞%"]/df["白细胞计数"]
    df["白淋细胞比"]=df["淋巴细胞%"]/df["白细胞计数"]
    df["白单细胞比"]=df["单核细胞%"]/df["白细胞计数"]
    df["白酸细胞比"]=df["嗜酸细胞%"]/df["白细胞计数"]
    df["白碱细胞比"]=df["嗜碱细胞%"]/df["白细胞计数"]
    df["中淋细胞比"]=df["中性粒细胞%"]/df["淋巴细胞%"]
    df["中单细胞比"]=df["中性粒细胞%"]/df["单核细胞%"]
    df["中酸细胞比"]=df["嗜酸细胞%"]/df["中性粒细胞%"]
    df["中碱细胞比"]=df["嗜碱细胞%"]/df["中性粒细胞%"]
    df["淋核细胞比"]=df["淋巴细胞%"]/df["单核细胞%"]
    df["淋酸细胞比"]=df["嗜酸细胞%"]/df["淋巴细胞%"]
    df["淋碱细胞比"]=df["嗜碱细胞%"]/df["淋巴细胞%"]
    df["核酸细胞比"]=df["嗜酸细胞%"]/df["单核细胞%"]
    df["核碱细胞比"]=df["嗜碱细胞%"]/df["单核细胞%"]
 
    #糖化血红蛋白构造
    df["血红蛋白/红细胞计数"]=df["血红蛋白"]/df["红细胞计数"]
    #红细胞各大指标比值
    df["红细胞压积/红细胞计数"]=df["红细胞压积"]/df["红细胞计数"]
    df["红细胞平均体积/红细胞计数"]=df["红细胞平均体积"]/df["红细胞计数"]
    df["红细胞平均血红蛋白量/红细胞计数"]=df["红细胞平均血红蛋白量"]/df["红细胞计数"]
    df["红细胞平均血红蛋白浓度/红细胞计数"]=df["红细胞平均血红蛋白浓度"]/df["红细胞计数"]
    df["红细胞体积分布宽度/红细胞计数"]=df["红细胞体积分布宽度"]/df["红细胞计数"]
    df["红细胞平均体积/血红蛋白"]=df["红细胞平均体积"]/df["血红蛋白"]  
    df["红细胞平均体积/红细胞压积"]=df["红细胞平均体积"]/df["红细胞压积"]  
    df["红细胞平均体积/红细胞平均血红蛋白浓度"]=df["红细胞平均体积"]/df["红细胞平均血红蛋白浓度"]  
    df["红细胞平均体积/红细胞平均血红蛋白量"]=df["红细胞平均体积"]/df["红细胞平均血红蛋白量"]  
    df["红细胞平均体积/红细胞体积分布宽度"]=df["红细胞平均体积"]/df["红细胞体积分布宽度"]  
 
    #血小板各大指标比值
    df["血小板计数/血小板平均体积"]=df["血小板计数"]/df["血小板平均体积"]
    df["血小板计数/血小板体积分布宽度"]=df["血小板计数"]/df["血小板体积分布宽度"]  
    df["血小板计数/血小板比积"]=df["血小板计数"]/df["血小板比积"]   
    df["血小板平均体积/血小板体积分布宽度"]=df["血小板平均体积"]/df["血小板体积分布宽度"]  
    df["血小板平均体积/血小板比积"]=df["血小板平均体积"]/df["血小板比积"]  
    df["血小板体积分布宽度/血小板比积"]=df["血小板体积分布宽度"]/df["血小板比积"]  
    
    #酶
    df["alt/ast"]=df["*天门冬氨酸氨基转换酶"]/df["*丙氨酸氨基转换酶"]  
    
    #蛋白肌酐比
    df["蛋白肌酐比"]=df["白蛋白"]/df["肌酐"]
            
    return df    
```
