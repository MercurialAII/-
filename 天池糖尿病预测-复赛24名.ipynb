```python
def insight(_):
    print(_,"的统计量分析")
    print(train[_].describe())
    null_sum=train[_].isnull().values.sum()
    if null_sum==0:
        print("没有缺失值")
    else:
        print("缺失值数量为:",null_sum,"缺失值比例为:" , null_sum/train.count().max())
    print("相关性：",train[_].corr(train["label"]))
    if len(train[_].unique())<10:
        print("离散变量:",train[_].groupby(train[_]).count())
    else:
        print("连续变量")
    plt.plot(train[_],train["label"],"o")
    plt.title("train")
    plt.show()
    sns.jointplot(train[_], train["label"], kind="hex", stat_func=kendalltau, color="#4CB391")
    plt.show()
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(1, figsize=(5, 5))#可设定图像大小
    p = train[[_]].boxplot() 
    plt.show()
```

```python
for _ in float_col:
    insight(_)
```

```python
#暴力特征之加减乘除log指数多项式
def make_feature(df):
    #暴力特征
    col_1=df.drop(["id","label"],axis=1).columns.tolist()
    col_2=df.drop(["id","label"],axis=1).columns.tolist()
    for _ in tqdm(col_1):
        df[_,"2"]=df[_]*df[_]
        col_2.remove(_)
        for __ in col_2:
            #df[_,"+",__]=df[_]+df[__]
            #df[_,"-",__]=df[_]-df[__]
            df[_,"*",__]=df[_]*df[__]
            df[_,"/",__]=df[_]/df[__]
    #提取log指数特征
    #col_list=df.drop(["id","label"],axis=1).columns.tolist()
    #for _ in tqdm(col_list):
    #    df["log",_]=np.log(df[_])
    #    df[_,"2"]=df[_]*df[_]
```

```python
make_feature(train)
testb["label"]=1
make_feature(testb)
testb.drop("label",axis=1,inplace=True)
```

```python
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
from sklearn import cross_validation, metrics
```

```python
lgb_model=lgb.LGBMClassifier(objective="binary",n_estimators=300,learning_rate =0.08,max_depth=3)
lgb_model.fit(train.drop(["id","label"],axis=1), train["label"])
lgb_pred=lgb_model.predict(testb.drop("id",axis=1))
```

```python
xgb_model=xgb.XGBClassifier(objective='binary:logistic',learning_rate =0.05,max_depth=5,n_estimators=100)
xgb_model.fit(train.drop(["id","label"],axis=1), train["label"])
xgb_pred=xgb_model.predict(testb.drop("id",axis=1))
```
